## Dataframe merge, concatenation and Groupby

```python
# In Pandas we merge with the merge method, which works pretty much like an SQL join with different tables
```

```python
# NOTE: Use merge, which is an inner join by default:
# join, which is a left join by default:
# concat, which is an outer join by default:
```

```python
df = pd.read_csv("train.csv") # importing .csv file
```

```python
df1 = df.head(3)
```

```python
df2 = df.tail(3)
```

```python
df1
```

```python
df2
```

```python
dfc = pd.concat([df1 , df2], axis = 0)  # concat data if axis is 0 so data comes under it
```

```python
dfc
```

```python
df3 = pd.concat([df1 , df2], axis = 1)  # it adds different rows
```

```python
df3
```

```python
df1 = pd.DataFrame({'a':range(6),
                    'b':[5,3,6,9,2,4]}, index=list('abcdef'))
```

```python
df2 = pd.DataFrame({'c':range(4),
                    'a':range(4),
                    'd':[10,20,30, 40]}, index=list('abhi'))
```

```python
df1
```

```python
df2
```

```python
# firstly use merge 
df3 = pd.merge(df1 , df2)   #inner join by default
```

```python
df3
```

```python
df3 = pd.merge(df1 , df2 ,how = "outer" , right_on="a" , left_on="a")  # right_on is the key for right df and same as left_on
```

```python
df3
```

```python
df3 = pd.merge(df1 , df2 ,how = "left" , right_on="a" , left_on="a")
```

```python
df3
```

```python
df3 = pd.merge(df1 , df2 ,how = "right" , right_on="a" , left_on="a") # it work like sql joins
```

```python
df3
```
