```python
df1 = pd.DataFrame({'a':range(6),
                    'b':[5,3,6,9,2,4]}, index=list('abcdef'))
```

```python
df2 = pd.DataFrame({'c':range(4),
                    'a':range(4),
                    'd':[10,20,30, 40]}, index=list('abhi'))
```

```python
df1
```

```python
df2
```

```python
# firstly use merge 
df3 = pd.merge(df1 , df2)   #inner join by default
```

```python
df3
```

```python
df3 = pd.merge(df1 , df2 ,how = "outer" , right_on="a" , left_on="a")  # right_on is the key for right df and same as left_on
```

```python
df3
```

```python
df3 = pd.merge(df1 , df2 ,how = "left" , right_on="a" , left_on="a")
```

```python
df3
```

```python
df3 = pd.merge(df1 , df2 ,how = "right" , right_on="a" , left_on="a") # it work like sql joins
```

```python
df3
```

```python
df4 = df3.groupby("d")  # similar to sql
```

```python
df4.c.mean()
```

## Apply, Map, and Applymap

```python
# Imagine you want to pass all the data belonging to one column through an specific function that transforms it in some way.
```

```python
# map: It is very useful for example when we want to convert a categorical value to a boolean one
```

```python
df3["pass"] = df3["d"].map(lambda marks : "Yes" if marks >= 33 else "No")
```

```python
# lets convert sex in boolean male for 1 and female for 0 this could have been done passing a dictionary as an argument, 
# where the keys represent the initial categories, and the values represent the new boolean levels
```

```python
df["sex_bool"] = df["Sex"].map({"male":1 , "female":0})
```

```python
df
```

```python
# s.apply(func, convert_dtype=True, args=()) syntax
```

```python
# lets convert Fare value doller to euro using apply method
```

```python
#define a function that converts dollars to euro
def dolar_to_euro(dollars, exchange_rate):
    euros = dollars*exchange_rate
    return euros
```

```python
df["Fare_Euro"] = df["Fare"].apply(dolar_to_euro , exchange_rate = 0.9)
```

```python
df
```

```python
# check old or not
```

```python
df["old_or_not"] = df["Age"].apply(lambda age : 1 if age > 40 else 0)
```

```python
df
```

```python
df.head(7)
```

```python
# to find women who is old
```

```python
df["old_women"] = df.apply(lambda row : 1 if (row["sex_bool"] == 0 and row["old_or_not"] == 1)  else 0 , axis = 1) 
```

```python
# we can create new dataframe where we have only data to old women
```

```python
owdf = df[df["old_women"] == 1]
```

```python
# ApplyMap is used when we have to apply on every field of an dataframe
import numpy as np
```

```python
#Creating this new dataframe
part_cols = ['Screw', 'Nut', 'Nail', 'Bearing']
duration_part_df = pd.DataFrame(np.random.randint(0,100,size=(100, 4)), columns=part_cols)
```

```python
duration_part_df
```

```python
# these data is for an perticular hour so if we need to convert it to whole year so we do
```
